```python
import pandas as pd
import requests
from urllib.parse import urljoin
```

```python
BASE_URL = "https://fantasy.premierleague.com/api/entry"
```

```python
manager_id = "5252797"
```

```python
MANAGER_URL = urljoin(f"{BASE_URL}/", manager_id)
```

```python
response = requests.get(MANAGER_URL)
```

```python
data = response.json()
```
